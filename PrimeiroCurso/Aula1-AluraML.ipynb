```python
# Ã© galinha (0) ou passaro (1)?
# tem crista?
# tem penas?
# voa?
galinha1 = [1,1,0]
galinha2 = [1,1,1]
galinha3 = [1,0,0]
passaro1 = [0,1,1]
passaro2 = [0,1,1]
passaro3 = [0,0,0]

treino_x = [galinha1, galinha2, galinha3, passaro1, passaro2, passaro3]
treino_y = [0,0,0,1,1,1]
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(treino_x, treino_y)

```

```python
from sklearn.metrics import accuracy_score

accuracy_score(model.predict([[1,0,0], [1,0,1]]), [1,0])
```
